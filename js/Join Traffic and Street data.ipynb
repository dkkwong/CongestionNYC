```python
traffic=gpd.read_file("C:/Users/David/Desktop/GEOG778/CongestionNYC/data/Traffic_Volume_Counts.csv")
traffic
```

```python
streets=gpd.read_file("C:/Users/David/Desktop/GEOG778/CongestionNYC/data/NYCStreets/NYCStreets.shp")
streets.plot()
```

```python
#merge dawta using segment ID
streetsMerge=traffic.merge(streets, on='SegmentID')
#remove geometry from traffic count table
streetsMerge.drop('geometry_x', axis=1, inplace=True)
```

```python
streetsMerge=gpd.GeoDataFrame(streetsMerge,geometry='geometry_y')
#Web Mercator
streetsMerge.to_crs(epsg=3857,inplace=True)
streetsMerge
```

```python
streetsMerge.to_file('C:/Users/David/Desktop/GEOG778/CongestionNYC/data/Streets.geojson', driver='GeoJSON') 
```
